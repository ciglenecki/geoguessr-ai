```python
import csv

csv_filepath = "coords_sample__n_1000000.csv"

coords = []

with open(csv_filepath) as f:
    reader = csv.reader(f, delimiter=',')
    next(reader)    # skip header
    for row in reader:
        coords.append(row)
```

```python
import requests

def get_metadata(coord, radius):
    meta_base = 'https://maps.googleapis.com/maps/api/streetview/metadata?'
    meta_params = {
        'key': api_key,
        'location': f"{coord[2]},{coord[1]}",
        'radius': radius,
        'return_error_code': 'true',
    }
    meta_response = requests.get(meta_base, params=meta_params)
    return meta_response.json()
```

```python
pano_ids = set()

results = []

dataset = coords[:1000]

try:
    for coord in dataset:
        response = get_metadata(coord = coord, radius=1000)

        if response.get('status') == 'OK':
            results.append([
                coord[0],
                coord[1],
                coord[2],
                True,
                response['location']['lng'],
                response['location']['lat'],
                response['pano_id'],
            ])
        elif (not response.get('status')) or (response.get('status') == 'ZERO_RESULTS'):
            results.append([
                coord[0],
                coord[1],
                coord[2],
                False,
                None,
                None,
                None,
            ])
        else:
            print(response)
except KeyboardInterrupt:
    last_coord_idx = dataset.index(coord)
    with open("results.csv" , 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(('id', 'long', 'lat', 'is_found', 'true_long', 'true_lat', 'pano_id'))
        for row in results:
            writer.writerow(row)
        
```
